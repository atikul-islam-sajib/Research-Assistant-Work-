```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def generate_boxplots(df, shrink_modes=None):
    # Customize Seaborn style and palette
    sns.set(style="whitegrid")

    # Define a custom color palette with attractive colors
    custom_palette = ["#3498db", "#e74c3c", "#2ecc71", "#f1c40f", "#9b59b6"]

    # Melt the DataFrame to convert it into a long format
    melted_df = pd.melt(df, id_vars=['relevance', 'shrink_mode'],
                        value_vars=['MDI_0', 'MDI_1', 'MDI_2', 'MDI_3', 'MDI_4'],
                        var_name='MDI', value_name='value')

    # Filter data based on the provided shrink_modes
    if shrink_modes:
        melted_df = melted_df[melted_df['shrink_mode'].isin(shrink_modes)]

    # Get the unique 'shrink_mode' values for the entire DataFrame
    unique_shrink_modes = melted_df['shrink_mode'].unique()

    # Define custom x-axis labels with LaTeX formatting
    x_labels = ['$X_1$', '$X_2$', '$X_3$', '$X_4$', '$X_5$']

    # Define the height of the FacetGrid plots
    plot_height = 6

    # Create a grid of boxplots with subplots in one row
    g = sns.FacetGrid(data=melted_df, col='relevance', sharey=True, height=plot_height, aspect=2,
                      col_wrap=len(melted_df['relevance'].unique()))

    # Customize the color palette with custom colors
    g.map_dataframe(sns.boxplot, x='MDI', y='value', hue='shrink_mode', hue_order=unique_shrink_modes,
                    palette=custom_palette)

    # Customize x-axis labels with LaTeX formatting
    for ax in g.axes.flat:
        ax.set_xticklabels(x_labels)

    # Customize plot appearance
    g.set_axis_labels('MDI', 'Value')
    g.despine(left=True)

    # Set the titles to include the relevance number
    for i, ax in enumerate(g.axes):
        ax.set_title(f'Relevance: {melted_df["relevance"].unique()[i]}')

    # Add legend to the last plot
    g.add_legend()

    return g

def grid_boxplots(df, relevance_values):
    # Create subplots for the specified relevance values
    for relevance in relevance_values:
        # Filter the DataFrame based on the current relevance value
        subset_df = df[df['relevance'] == relevance]

        # Call the generate_boxplots function with the filtered DataFrame
        g = generate_boxplots(subset_df, shrink_modes=['hs', 'hs_entropy'])

        # Save each grid as a separate PDF
        g.savefig(f'boxplots_relevance_{relevance}.pdf', bbox_inches='tight')

# Example usage: grid_boxplots(df, relevance=[0, 0.1, 0.2])

# Create subplots for the specified relevance values and save them as separate PDFs
grid_boxplots(df, relevance_values=[0, 10, 20])

# Show the plots
plt.show()
```
